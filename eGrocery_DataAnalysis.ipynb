```python
# Headers
import pandas as pd
```

```python
# Data Loading
data_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/eGrocery_Project/final_datax.csv')
```

```python
products = [8,11,15,17,94,95,95,110,112,128]
prod_count = len(products)
columns = ['prod'+str(prod) for prod in products]
print(columns)
```

```python
#Plotting demands of all 10 products for 761 days
import matplotlib.pyplot as plt

demands = [data_df[col].tolist() for col in columns]
indices = [i for i in range(len(demands[0]))]
for i,col in enumerate(columns):
  print(col)
  plt.scatter(indices,demands[i])
  plt.show()
```

```python
import numpy as np

demands_yesterday = [demands[i][:len(demands[i])-1] for i in range(prod_count)]
demands_2days_back = [demands[i][:len(demands[i])-2] for i in range(prod_count)]
demands_3days_back = [demands[i][:len(demands[i])-3] for i in range(prod_count)]
demands_4days_back = [demands[i][:len(demands[i])-4] for i in range(prod_count)]
demands_5days_back = [demands[i][:len(demands[i])-5] for i in range(prod_count)]
demands_6days_back = [demands[i][:len(demands[i])-6] for i in range(prod_count)]
demands_lastweek = [demands[i][:len(demands[i])-7] for i in range(prod_count)]

print('Correlation of todays demand with yesterdays demand')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][1:]),np.asarray(demands_yesterday[i]))[0][1])

print('Correlation of todays demand with demand 2 days back')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][2:]),np.asarray(demands_2days_back[i]))[0][1])

print('Correlation of todays demand with demand 3 days back')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][3:]),np.asarray(demands_3days_back[i]))[0][1])

print('Correlation of todays demand with demand 4 days back')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][4:]),np.asarray(demands_4days_back[i]))[0][1])

print('Correlation of todays demand with demand 5 days back')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][5:]),np.asarray(demands_5days_back[i]))[0][1])
print('Correlation of todays demand with demand 6 days back')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][6:]),np.asarray(demands_6days_back[i]))[0][1])

print('Correlation of todays demand with previous same weekday demand')
for i,col in enumerate(columns):
  print(col,':',np.corrcoef(np.asarray(demands[i][7:]),np.asarray(demands_lastweek[i]))[0][1])
```

The correlation scores sort of makes sense as the correlation decreases as we go back in the week. Its interesting to see that correlation still goes down with demand of last same weekday.
